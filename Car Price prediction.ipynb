## Correlation

```python
for i in num[:-1]:
    plt.figure(figsize=(6,3))

    coeff = np.polyfit(df[f'{i}'], df.price, 1)
    f = np.poly1d(coeff)
    plt.scatter(df[f'{i}'], df.price)
    plt.plot(df[f'{i}'], f(df[f'{i}']))
    plt.title('correlation = '+ str(df.price.corr(df[f'{i}'])))
    plt.xlabel(f'{i}')
    plt.ylabel('Price')
    plt.show()
```

```python
## ployfit practice
coeff = np.polyfit(df.highwaympg, df.price, 1)
f = np.poly1d(coeff)
plt.scatter(df.highwaympg, df.price)
plt.plot(df.highwaympg, f(df.highwaympg))
```
