```python
md['Sales_sqrt'] = md.Sales.map(lambda x: np.sqrt(x) if x != 0 else 0)
```

```python
md['Customers_ln'] = np.log(md.Customers)
```

```python
md['CompetitionDistance_ln'] = np.log(md.CompetitionDistance)
```

```python
md['PromoOpen_ln'] = md.promo_open.map(lambda x: np.log(x) if x != 0 else 0)
```

```python
md['CompetitionOpen_ln'] = md.competition_open.map(lambda x: np.log(x) if x != 0 else 0)
```

```python
md.drop(['Sales_ln'],axis = 1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
Scale = StandardScaler()
data = md[['Sales_sqrt', 'CompetitionOpen_ln', 'PromoOpen_ln', 'CompetitionDistance_ln', 'Customers_ln']]
std_data = Scale.fit_transform(data)
std_data = pd.DataFrame(std_data, columns = 'Scaled_'+data.columns)
std_data.tail()
```

```python
std_data.isnull().sum()
```

```python
md2 = pd.concat([md, std_data], axis = 1)
```

```python
md2.columns
```

```python
md2.transpose()
```

```python
md_final = md2[['DayOfWeek',  'Open',
       'Promo', 'SchoolHoliday', 'year', 'month', 'day',
       'StoreType_b', 'StoreType_c', 'StoreType_d',
       'Assortment_b', 'Assortment_c', 'StateHoliday_0', 'StateHoliday_a',
       'StateHoliday_b', 'StateHoliday_c', 'PromoInterval_Jan,Apr,Jul,Oct',
       'PromoInterval_Mar,Jun,Sept,Dec', 'Scaled_Sales_sqrt', 'Scaled_CompetitionOpen_ln',
       'Scaled_PromoOpen_ln', 'Scaled_CompetitionDistance_ln',
       'Scaled_Customers_ln']]
```

## Linear Regression

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import *
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_selection import RFE
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
x = md_final.drop(['Scaled_Sales_sqrt'], axis = 1)
```

```python
y = md_final[['Scaled_Sales_sqrt']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
